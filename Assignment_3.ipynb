```python
# LGIS_EX
def read_file(filename):
    with open(filename, 'r') as file:
        n = int(file.readline().strip())
        perm = list(map(int, file.readline().strip().split())) 
    
    return perm 

def find_increasing_subsequence(perm):
    n = len(perm)
    lengths = [1]*n             # lengths of LIS
    previuos_index = [-1] * n   # tracking the previous element in LIS

    for i in range(1,n):
        for j in range(i): # inner loop that iterates through all indices before i
            if perm[i] > perm[j] and lengths[i] < lengths[j] + 1:
                lengths[i] = lengths[j] + 1
                previuos_index[i] = j

    # Finding the LIS:
    lis_length = max(lengths)
    lis_end = lengths.index(lis_length)
    lis = []

    while lis_end != -1:
        lis.append(perm[lis_end])
        lis_end = previuos_index[lis_end] 

    lis.reverse() # reversing the list to get the correct order
    return lis
    
def find_decreasing_subsequence(perm):
        return find_increasing_subsequence(perm[::-1])[::-1] # reversing the result to get LDS
    
def permutation(perm):
        lis = find_increasing_subsequence(perm)
        lds = find_decreasing_subsequence(perm)

        return lis, lds
    
filename = 'rosalind_lgis.txt'
perm = read_file(filename)
lis, lds = permutation(perm)
print(" ".join(map(str,lis)))
print(" ".join(map(str, lds)))
```

```python
# EDIT_EX

def read_fasta(file_path):
    with open(file_path, 'r') as f:
        lines = f.readlines()
    
    sequences = []
    current_sequence = []
    for line in lines:
        if line.startswith(">"):
            if current_sequence:
                sequences.append("".join(current_sequence))
                current_sequence = []
        else:
            current_sequence.append(line.strip())
    if current_sequence:
        sequences.append("".join(current_sequence))
    return sequences

# Calculating the edit distance between the two strings: 
def edit_distance(s, t):

    m, n = len(s), len(t)
    dp = [[0] * (n + 1) for i in range(m + 1)]
    
    # Base case initialization: 
    for i in range(m + 1): 
         dp[i][0] = i
    for j in range(n + 1):
        dp[0][j] = j
    
    for i in range(1, m + 1):
        for j in range(1, n + 1):
            if s[i - 1] == t[j - 1]:
                dp[i][j] = dp[i - 1][j - 1]   # no operations are needed
            else:
                dp[i][j] = 1 + min(dp[i - 1][j], dp[i][j - 1], dp[i - 1][j - 1])
                                    # deletion   # insertion    # substitution
    return dp[m][n]

input_file = "rosalind_edit.fasta"  
sequences = read_fasta(input_file)
if len(sequences) >= 2:
    s, t = sequences[0], sequences[1]
    result = edit_distance(s, t)
    print(result)
else:
    print("Error: The input file must contain at least 2 sequences")
```

```python
# EDTA_EX

def edit_distance(s, t):
    m, n = len(s), len(t)
  
    dp = [[0] * (n + 1) for _ in range(m + 1)]
    
   
    for i in range(m + 1):
        dp[i][0] = i 
    for j in range(n + 1):
        dp[0][j] = j  

    # Filling the DP table
    for i in range(1, m + 1):
        for j in range(1, n + 1):
            if s[i - 1] == t[j - 1]:
                dp[i][j] = dp[i - 1][j - 1]  # No cost for matching characters
            else:
                dp[i][j] = min(
                    dp[i - 1][j - 1] + 1,  # Substitution
                    dp[i][j - 1] + 1,      # Insertion
                    dp[i - 1][j] + 1       # Deletion
                )

    # Backtracking to find the optimal alignment:
    aligned_s = []
    aligned_t = []
    i, j = m, n

    while i > 0 or j > 0:
        if i > 0 and j > 0 and (s[i - 1] == t[j - 1] or dp[i][j] == dp[i - 1][j - 1] + 1): # checking if the characters at s[i-1] and t[j-1] are either the same or if the current cell in the DP table (dp[i][j]) is the result of a substitution
            aligned_s.append(s[i - 1])  
            aligned_t.append(t[j - 1])
            i -= 1
            j -= 1
        elif i > 0 and dp[i][j] == dp[i - 1][j] + 1:
            aligned_s.append(s[i - 1])
            aligned_t.append('-')      # appending gap (-) to the aligned sequence for t to represent the deletion.
            i -= 1                     # decrementing i to move one step left in s to align with the current position in t.
        else:
            aligned_s.append('-')      
            aligned_t.append(t[j - 1]) # gap (-) added to the aligned sequence for s to represent the insertion. 
            j -= 1                     # decrementing j to move one step left in t, indicating that we've inserted t[j-1] into s.

    # Reversing aligned sequences:
    aligned_s.reverse()
    aligned_t.reverse()

    
    return dp[m][n], ''.join(aligned_s), ''.join(aligned_t)


def read_fasta(file_path):
    with open(file_path, 'r') as file:
        lines = file.readlines()
        seqs = []
        current_seq = ""
        for line in lines:
            if line.startswith(">"): 
                if current_seq:  
                    seqs.append(current_seq)
                current_seq = ""  
            else:
                current_seq += line.strip()  
        seqs.append(current_seq)  
        return seqs[0], seqs[1]  


file_path = "rosalind_edta.fasta"  
s, t = read_fasta(file_path)

edit_distance_value, aligned_s, aligned_t = edit_distance(s, t)

print(edit_distance_value)
print(aligned_s)
print(aligned_t)
```

```python
# CTEA_EX

def read_fasta(filename):
    with open(filename, 'r') as file:
        lines = file.readlines()

    sequences = []
    current_sequence = []

    for line in lines:
        line = line.strip()
        if line.startswith('>'):
            if current_sequence:
                sequences.append(''.join(current_sequence))  
            current_sequence = [] 
        else:
            current_sequence.append(line)  
    if current_sequence:
        sequences.append(''.join(current_sequence))  

    return sequences

# Computing the minimum edit distance and the number of optimal alignments
def optimal_alignments(s, t, mod=134217727):
    m, n = len(s), len(t)
    
    # Initializing DP tables:
    dp = [[0] * (n + 1) for _ in range(m + 1)]  # edit distance
    count = [[0] * (n + 1) for _ in range(m + 1)]  # counting optimal alignments
    
    for i in range(m + 1):
        dp[i][0] = i
        count[i][0] = 1
    for j in range(n + 1):
        dp[0][j] = j
        count[0][j] = 1
    
    for i in range(1, m + 1):
        for j in range(1, n + 1):
            # Cost of substitution or matching
            substitution_cost = dp[i - 1][j - 1] + (0 if s[i - 1] == t[j - 1] else 1)
            
            # Cost of insertion and deletion
            insertion_cost = dp[i][j - 1] + 1
            deletion_cost = dp[i - 1][j] + 1
            
            # Taking the minimum cost:
            min_cost = min(substitution_cost, insertion_cost, deletion_cost)
            dp[i][j] = min_cost
            
            # Counting how many ways there are to achieve the minimum edit distance:
            if dp[i - 1][j - 1] + (0 if s[i - 1] == t[j - 1] else 1) == min_cost: # through substitution or no operation if the char match
                count[i][j] += count[i - 1][j - 1]
            if dp[i][j - 1] + 1 == min_cost:                                      # through insertion
                count[i][j] += count[i][j - 1]
            if dp[i - 1][j] + 1 == min_cost:                                      # through deletion
                count[i][j] += count[i - 1][j]
            
            count[i][j] %= mod  # ensuring that the count doesn't exceed a certain limit. 
    
    # The number of optimal alignments is in count[m][n]
    return count[m][n]

if __name__ == "__main__":
    
    filename = "rosalind_ctea.fasta"  
    sequences = read_fasta(filename)
    
    if len(sequences) == 2:
        s = sequences[0]
        t = sequences[1]
        result = optimal_alignments(s, t)
        print(result)
    else:
        print("The FASTA file must contain exactly two sequences.")
```

```python
# BIOPYTHON: 
# Add your functions here

from Bio import pairwise2
from Bio import SeqIO

def countMatches(a, b):
    matches = 0
    for element in range(len(a)):  
        if a[element] == b[element] and a[element] != '-' and b[element] != '-':  
            matches += 1
    return matches

def countMismatches(a, b):
    mismatches = 0 
    for element in range(len(a)):
        if a[element] != b[element] and a[element] != '-' and b[element] != '-':
            mismatches += 1
    return mismatches

def countGapOpens(a, b):
    gap_opens = 0
    if a[0] == '-' and b[0] == '-':
        gap_opens += 1
    for element in range(1, len(a)):  
        if (a[element] == '-' and a[element - 1] != '-') or (b[element] == '-' and b[element - 1] != '-'):
            gap_opens += 1
    return gap_opens

def countGapExtensions(a, b):
    gap_extensions = 0
    for element in range(1, len(a)):
        if (a[element] == '-' and a[element - 1] == '-') or (b[element] == '-' and b[element - 1] == '-'):
            gap_extensions += 1
    return gap_extensions


def getScore(a, b, matchScore, mismatchPenalty, gapOpenPenalty, gapExtensionPenalty):
    matches = countMatches(a, b)
    mismatches = countMismatches(a, b)
    gap_opens = countGapOpens(a, b)
    gap_extensions = countGapExtensions(a, b)

    total_score = (   
        matches * matchScore +
        mismatches * mismatchPenalty +
        gap_opens * gapOpenPenalty +
        gap_extensions * gapExtensionPenalty
    )
    return total_score

def read_sequence(file_path):
    record = SeqIO.read(file_path, "fasta")
    return record.seq

a = read_sequence("IL12A.fasta")
b = read_sequence("IL12B.fasta")

# Global alignment with pairwise:
alignments = pairwise2.align.globalms(a, b, 1, -1, -0.5, -0.2)
pairwise_alignment = alignments[0]
aligned_a, aligned_b = pairwise_alignment.seqA, pairwise_alignment.seqB

# Scoring parameters
matchScore = 1
mismatchPenalty = -1
gapOpenPenalty = -0.5
gapExtensionPenalty = -0.2

# Alignment score components: 
matches = countMatches(aligned_a, aligned_b)
mismatches = countMismatches(aligned_a, aligned_b)
gap_opens = countGapOpens(aligned_a, aligned_b)
gap_extensions = countGapExtensions(aligned_a, aligned_b)

# Final score: 
score = getScore(aligned_a, aligned_b, matchScore, mismatchPenalty, gapOpenPenalty, gapExtensionPenalty)

print(f"The alignemnt score is: {score}")           
print(f"\nThe pairwise alignemnt score is: {pairwise_alignment.score}")   
```
